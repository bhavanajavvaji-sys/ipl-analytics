## Notebook 1 - Data Loading & CLeaning
**Goal:** Load the dataset, understand its structure, fix data qualtiy issues and prepare it for analysis. 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

print("Libraries loaded successfully")
print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
```

```python
df = pd.read_csv('../data/IPL.csv')

print(f"data loaded successfully")
print(f"Rows : {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
print(f"Total data points: {df.shape[0]*df.shape[1]:,}")
```

```python
print("=== FIRST 3 ROWs ===")
df.head(3)
```

```python
print("=== COLUMN INFORMATION ===")
print(f"{'column':<25} {'Non-Null': <15} {'Dtype'}")
print("-"* 50)
for col in df.columns:
    non_null = df[col].notna().sum()
    dtype = df[col].dtype
    print(f"{col:<25} {non_null:<15} {dtype}")
```

```python
#missing value summary
print("=== MISSING VALUE REPORT ===\n")

missing = df.isnull().sum()
missing_pct = (missing /len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'Missing Count': missing,
    'Missing %' : missing_pct
})

#Only show columns that have missing values
missing_report = missing_report[missing_report['Missing Count'] > 0 ]
missing_report = missing_report.sort_values('Missing %' , ascending=False)


print(missing_report)
print(f"\nTotal colums with missing data: {len(missing_report)}")
print(f"\nColums fully complete: {df.shape[1] - len(missing_report)}")

```

```python
# Fix date column
df['date'] = pd.to_datetime(df['date'])

# Usefull time components
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

print("Date column fixed")
print(f"Date dtype now : {df['date'].dtype}")
print(f"\nSeasons in data : {sorted(df['year'].unique())}")
print(f"\nTotal Seasons : {df['year'].nunique()}")
```

```python
# Standardize team names
team_name_fix = {
    'Delhi Daredevils' : 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Pune Warriors': 'Rising Pune Supergiant',
    'Kings XI Punjab' : 'Punjab Kings',
    'Rising Pune Supergiants' : 'Rising Pune Supergiant',
    'Royal Challengers Bengalore' : 'Royal Challengers Bengalure'
    }

#applying to all revelent columns 
for col in ['batting_team' , 'bowling_team']:
    df[col] = df[col].replace(team_name_fix)

#verify
print("Team names standardized")
print(f"\nUnique teams now:")
for team in sorted(df['batting_team'].unique()):
    print(f" {team}")
```

```python
# fix remaining name issues
extra_fixes = {
    'Royal Challengers Bengalure' : 'Royal Challengers Bengaluru'
}

for col in ['batting_team', 'bowling_team']:
    df[col] = df[col].replace(extra_fixes)

#verify count dropped
print(f"\nAfter fix - unique teams: {df['batting_team'].nunique()}")
print(df['batting_team'].unique())
```

```python
#check win_outcome before cleaning
print(f"Total rows before: {len(df):,}")
print(f"Missing win_outcome: {df['win_outcome'].isnull().sum():,}")

#keep only completed matches
df_clean = df.dropna(subset =['win_outcome']).copy()

print(f"\nTotal rows after: {len(df_clean):,}")
print(f"Rows removed: {len(df)-len(df_clean):,}")
```

```python
# save cleaned dataframe
df_clean.to_csv('../data/IPL_clean.csv', index = False)

print("clean data saved")
print(f"File saved to : data/IPL_clean.csv")
print(f"ROws: {len(df_clean):,}")
print(f"Colums: {df_clean.shape[1]}")
```
